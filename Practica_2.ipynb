**Practica 2**

El programa abre el archivo informe_acciones.txt, que contiene varios correos electrónicos con información sobre el valor de las acciones de distintas empresas, y utiliza expresiones regulares para identificar en cada mensaje el nombre de la empresa y el valor numérico asociado. Primero detecta la empresa a partir de la línea del Asunto y después busca en el cuerpo del correo la línea donde aparece la palabra “Valor” junto con el número decimal. Cada par de datos se guarda en una lista y se imprime en pantalla mostrando la empresa con su valor actual de acciones. Finalmente, el programa calcula estadísticas generales como el total de empresas procesadas, el valor mínimo, el valor máximo y el promedio de las acciones, ofreciendo así un resumen claro y completo del análisis realizado.



```python
import re

ruta = "/content/drive/MyDrive/ClassFiles/informe_acciones.txt"

acciones = []
empresa_actual = None

# capturar empresa en el Asunto
patron_empresa = re.compile(r"Asunto:\s*Informe de Acción\s*-\s*(\w+)", re.IGNORECASE)

# capturar el valor numérico en el cuerpo
patron_valor = re.compile(r"Valor.*?([\d]+\.\d+)", re.IGNORECASE)

with open(ruta, "r", encoding="latin-1") as f:
    for linea in f:
        linea = linea.strip()

        # decidi que es bueno indicar la empresa junto con su valor
        #Si la línea contiene el Asunto, guardamos la empresa
        match_empresa = patron_empresa.search(linea)
        if match_empresa:
            empresa_actual = match_empresa.group(1)

        # Si la línea contiene el valor, lo asociamos con la empresa actual
        match_valor = patron_valor.search(linea)
        if match_valor and empresa_actual:
            valor = float(match_valor.group(1))
            acciones.append((empresa_actual, valor))
            empresa_actual = None

print("Valores actuales de las acciones:\n")
if acciones:
    for empresa, valor in acciones:
        print(f"{empresa}: {valor}")

# Agregue esto extra para conocer el total de empresas y valores minimo y maximo

        total_empresas = len(acciones)
    valores = [v for _, v in acciones]
    valor_min = min(valores)
    valor_max = max(valores)
    valor_promedio = sum(valores) / len(valores)
    print("\nResumen:")
    print("Total de empresas procesadas:", total_empresas)
    print("Valor mínimo de las acciones:", valor_min)
    print("Valor máximo de las acciones:", valor_max)
    print("Valor promedio de las acciones:", valor_promedio)


else:
    print("No se encontraron coincidencias. Revisa el formato de las líneas en tu archivo.")

f.close()
```
