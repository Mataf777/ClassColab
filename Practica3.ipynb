Creación de un arreglo y operaciones básicas.

```python
import numpy as np

arreglo = np.random.randint(1, 101, size=20)
arreglo

# Valor máximo del arreglo
maximo = arreglo.max()

# Valor mínimo del arreglo
minimo = arreglo.min()

# Media (promedio) del arreglo
media = arreglo.mean()

# Sumar 10 a cada elemento del arreglo
arreglo_mas_10 = arreglo + 10

maximo, minimo, media, arreglo_mas_10

print("Arreglo original:")
print(arreglo)

print("\nResultados:")
print(f"  Máximo del arreglo = {maximo}")
print(f"  Mínimo del arreglo = {minimo}")
print(f"  Media del arreglo = {media:.2f}")

print("\nArreglo con 10 sumado a cada elemento:")
print(arreglo_mas_10)
```
