5. Analisis de estructura


```python
import pandas as pd
import pandas as pd
pd.set_option('display.float_format', '{:.3f}'.format)

df = pd.read_csv("/content/drive/MyDrive/ClassFiles/beisstats.csv")

df.shape
df.info()
df.isnull().sum()
df.describe()

```

4. Preguntas Descriptivas

     1.-Cual es la distribucion del wOBA (promedio de embase) entre los jugadores?


```python
import numpy as np

print("Media:", round(df['woba'].mean(), 3))
print("Mediana:", np.median(df['woba']))
print("Desviación estándar:", round(df['woba'].std(), 3))
print("Mínimo:", np.min(df['woba']))
print("Máximo:", np.max(df['woba']))
```

El wOBA(Promedio para embasarse ponderado) presenta una media de 0.308 y una mediana muy cercana, lo que indica una distribución simétrica y un rendimiento ofensivo estable entre los jugadores analizados. La desviación estándar moderada sugiere que, aunque el grupo es relativamente homogéneo, existen diferencias individuales relevantes que podrían explicarse por variaciones en la calidad de contacto o disciplina en el plato.

 2.-Cual es la media y desviacion estandar de hard_hit% (pelotas bateadas con velocidad de salida alta)



```python
print("Media hard_hit%:", round(df['hard_hit_percent'].mean(), 3))
print("Desviación estándar hard_hit%:", round(df['hard_hit_percent'].std(), 3))
print("Mínimo:", df['hard_hit_percent'].min())
print("Máximo:", df['hard_hit_percent'].max())

```

El hard_hit% presenta una media de 40.17%, lo que indica que el grupo analizado genera contacto fuerte con una frecuencia superior al promedio general de la liga. La desviación estándar moderada (4.08) sugiere una variabilidad controlada entre jugadores, aunque se observan diferencias importantes entre el valor mínimo (28.9%) y el máximo (50%), lo que indica que la capacidad de generar contacto fuerte puede ser un factor diferenciador clave en el rendimiento ofensivo.

 3.-Existe correlacion entre barrel%(batazo con potencial) y xwOBA (embase esperado)?



```python
from scipy.stats import pearsonr


r, p = pearsonr(df['barrel_batted_rate'], df['xwoba'])
print(f"Correlación Barrel% vs xwOBA: {r:.3f}")
print(f"p-value: {p:.3e}")
```

Se encontró una correlación positiva moderada (r = 0.498) entre barrel%(calidad de contacto del bateador) y xwOBA(Promedio de Embasarse Ponderado esperado), la cual es estadísticamente significativa (p < 0.001). Esto indica que los jugadores que generan una mayor proporción de batazos con alto potencial tienden a presentar mejores valores de rendimiento ofensivo esperado. Aproximadamente el 25% de la variabilidad del xwOBA puede explicarse por el barrel%, lo que confirma que la calidad del contacto es un factor relevante en la producción ofensiva

 4.-Que jugadores presentan mayor whiff% (abanicar ) y como se relaciona con k_percent (strikes)?



```python
df.sort_values(by='whiff_percent', ascending=False)[
    ['last_name, first_name','whiff_percent','k_percent']
].head(5)

from scipy.stats import pearsonr

r, p = pearsonr(df['whiff_percent'], df['k_percent'])
print(f"Correlacion Whiff vs K: {r:.3f}")
print(f"p-value: {p:.3e}")


```

Se encontró una correlación positiva muy fuerte entre whiff%(bateador abanica y falla ) y K%(ponches) (r = 0.866), estadísticamente significativa (p < 0.001). Esto indica que los jugadores que presentan mayores tasas de swings fallidos tienden a registrar significativamente más ponches. Aproximadamente el 75% de la variabilidad en el K% puede explicarse por el whiff%, lo que confirma que la frecuencia de abanicos es un indicador determinante del riesgo ofensivo.


 5.-Existe diferencia significative entre wOBA y xwOBA en promedio?


```python
from scipy.stats import ttest_rel

t_stat, p_value = ttest_rel(df['woba'], df['xwoba'])

print(f"Estadístico t: {t_stat:.3f}")
print(f"p-value: {p_value:.3e}")
```

La prueba *paired T test* mostró una diferencia estadísticamente significativa entre wOBA(promedio para embasarse ponderado) y xwOBA(Promedio de Embasarse Ponderado esperado) (t = -7.59, p < 0.001), lo que indica que el rendimiento ofensivo observado difiere del esperado según la calidad del contacto. El signo negativo del estadístico sugiere que, en promedio, los jugadores presentan un wOBA inferior a su xwOBA, lo que podría reflejar influencia de factores externos como variabilidad, defensa o efectos situacionales.
