#Practica 5

 Uso de máscaras booleanas.



```python
import numpy as np

# Crear arreglo unidimensional con 15 números aleatorios entre 1 y 50
arreglo = np.random.randint(1, 51, size=15)

print("Arreglo original:")
print(arreglo)

# Máscara booleana: elementos mayores que 25
mascara = arreglo > 25

print("\nMáscara booleana (arreglo > 25):")
print(mascara)

# Seleccionar elementos mayores que 25
mayores_25 = arreglo[mascara]

print("\nElementos mayores que 25:")
print(mayores_25)

# Contar cuántos elementos son mayores que 25
cantidad = mayores_25.size

print(f"\nCantidad de elementos mayores que 25 = {cantidad}")
```
