#Práctica 12: Carga de Datos

Al final del archivo contiene una línea que dice "Esta es una línea que no deberías cargar". Esta línea es irrelevante por lo tanto al cargar el CSV a un dataframe se debería de ignorar.

```python
import pandas as pd

# Ruta del archivo
ruta = "/content/drive/MyDrive/ClassFiles/miaad-nyc-r-s.csv"

# Cargar CSV omitiendo la última fila
df = pd.read_csv(ruta, skipfooter=1, engine="python")

# Mostrar las últimas 3 filas para comprobar que no se cargó la línea irrelevante
print("Últimas 3 filas:")
print(df.tail(3))

# Mostrar dimensiones del dataframe
print("\nDimensiones del dataframe:")
print("Filas:", df.shape[0])
print("Columnas:", df.shape[1])
```

#Práctica 13: Cambiar nombre de columna

Imprime los nombres de la columnas del archivo
Asegúrate que el nombre de la primera columna debería ser INDEX MIAAD. Si esto no es así, entonces cambia el nombre de la columna y vuelve a imprimir los nombres para mostrar el cambio realizado. Después del cambio debería aparecer INDEX MIAAD

```python

#Mostrar nombres originales

print("Nombres originales de las columnas:")
print(df.columns.tolist())


#Verificar y cambiar si es necesario


if df.columns[0] != "INDEX MIAAD":
    print("\nLa primera columna NO se llama 'INDEX MIAAD'. Se realizará el cambio...\n")

    # Cambiar nombre de la primera columna
    df.rename(columns={df.columns[0]: "INDEX MIAAD"}, inplace=True)


#Mostrar nombres después del cambio


print("Nombres de las columnas después de la verificación/cambio:")
print(df.columns.tolist())

```

#Práctica 14: Normalización de los nombres de la columna.

Normaliza los nombres de las columnas, los cuales deberán estar en minúsculas. Asimismo, reemplaza los espacios con guion bajo
Imprime los nombres de las columnas para demostrar que has realizado la normalización.

```python

#Mostrar nombres originales


print("Nombres originales:")
print(df.columns.tolist())


#Normalizar nombres


df.columns = (
    df.columns
    .str.lower()          # convertir a minúsculas
    .str.replace(" ", "_") # reemplazar espacios por _
)


#Mostrar nombres normalizados


print("\nNombres después de la normalización:")
print(df.columns.tolist())
```

#Práctica 15: Errores en los Tipos de Datos

Imprime los tipos de datos de cada columna.
Asegúrate que las columnas de SALE PRICE(sale_price), LAND SQUARE FEET (land_square_feet), GROSS SQUARE FEET (gross_square_feet) sean de tipo flotante. Escribe qué tipo de dato tienen asignado. Si no son correctos, realiza el cambio.
Imprime de nuevo los tipos de datos para demostrar que has arreglado el problema.

```python

#Ver tipos originales


print("Tipos de datos originales:\n")
print(df.dtypes)


#Verificar tipos específicos


print("\nTipos actuales de las columnas importantes:")
print("sale_price:", df["sale_price"].dtype)
print("land_square_feet:", df["land_square_feet"].dtype)
print("gross_square_feet:", df["gross_square_feet"].dtype)


#Convertir a tipo float si es necesario


df["sale_price"] = pd.to_numeric(df["sale_price"], errors="coerce")
df["land_square_feet"] = pd.to_numeric(df["land_square_feet"], errors="coerce")
df["gross_square_feet"] = pd.to_numeric(df["gross_square_feet"], errors="coerce")


#Verificar nuevamente


print("\nTipos de datos después de la conversión:\n")
print(df.dtypes)


print("\nTipos actuales de las columnas importantes:")
print("sale_price:", df["sale_price"].dtype)
print("land_square_feet:", df["land_square_feet"].dtype)
print("gross_square_feet:", df["gross_square_feet"].dtype)
```

#Práctica 16: Genera el diccionario de datos del dataframe preprocesado. Imprime el contenido de dicho diccionario

```python

#Crear diccionario de datos


diccionario_datos = {}

for col in df.columns:
    diccionario_datos[col] = {
        "tipo_de_dato": str(df[col].dtype),
        "valores_no_nulos": int(df[col].notnull().sum()),
        "valores_nulos": int(df[col].isnull().sum())
    }


#Imprimir el diccionario


print("Diccionario de datos del dataframe preprocesado:\n")

for columna, info in diccionario_datos.items():
    print(f"Columna: {columna}")
    print(f"  Tipo de dato: {info['tipo_de_dato']}")
    print(f"  Valores no nulos: {info['valores_no_nulos']}")
    print(f"  Valores nulos: {info['valores_nulos']}")
    print("-" * 40)
```
