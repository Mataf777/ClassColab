```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


# Cargar el dataset Iris desde seaborn
df = sns.load_dataset('iris')


# Explorar las primeras filas del dataset
print(df.head())



# Obtener informacin general del dataset
# (columnas, tipos de datos, valores nulos, etc.)
df.info()


# Obtener estadsticas descriptivas de las variables numricas
df.describe()

# a) Objetivo: Comparar el promedio de largo y ancho de los ptalos entre las tres especies de Iris.
# a) Calcular el promedio del largo y ancho del ptalo por especie
promedios = df.groupby('species')[['petal_length', 'petal_width']].mean()


# a) Preparar los datos para el grfico de barras
especies = promedios.index
petal_length_mean = promedios['petal_length']
petal_width_mean = promedios['petal_width']


# a) Definir posiciones y ancho de las barras
x = np.arange(len(especies))  # posiciones en el eje X
width = 0.35                  # ancho de cada barra


# a) Crear el grfico de barras
plt.figure(figsize=(8, 5))

plt.bar(x - width/2, petal_length_mean, width, label='Petal Length')
plt.bar(x + width/2, petal_width_mean, width, label='Petal Width')


# a) Personalizar el grfico
plt.xticks(x, especies)
plt.ylabel('Promedio')
plt.title('Promedio de largo y ancho del ptalo por especie')
plt.legend()


# a) Mostrar el grfico
plt.show()

# b) Objetivo: Visualizar la distribucin de los largos de los spalos (sepal_length)
# b) Histograma

# b) Seleccionar la variable sepal_length
sepal_length = df['sepal_length']

# b) Crear el histograma
plt.figure(figsize=(8, 5))
plt.hist(sepal_length, bins=20)

# b) Etiquetas y ttulo del grfico
plt.xlabel('Largo del spalo (sepal_length)')
plt.ylabel('Frecuencia')
plt.title('Distribucin del largo del spalo')

# b) Mostrar el histograma
plt.show()

# c) Objetivo: Relacionar el largo del ptalo con el ancho del ptalo
# c) Grfico de dispersin (scatter plot)
# c) usando un color diferente para cada especie

plt.figure(figsize=(8, 5))

# c) Obtener las especies nicas
especies = df['species'].unique()

# c) Crear un scatter por cada especie
for especie in especies:
    datos_especie = df[df['species'] == especie]
    plt.scatter(
        datos_especie['petal_length'],
        datos_especie['petal_width'],
        label=especie
    )

# c) Etiquetas y ttulo
plt.xlabel('Largo del ptalo (petal_length)')
plt.ylabel('Ancho del ptalo (petal_width)')
plt.title('Relacin entre largo y ancho del ptalo')

# c) Leyenda
plt.legend()

# c) Mostrar el grfico
plt.show()

# d) Grfico de cajas (box plot)
# d) Objetivo: Analizar la distribucin de las medidas del spalo
# d) (sepal_length y sepal_width) por especie

plt.figure(figsize=(10, 5))

# d) Crear boxplots para cada especie y cada variable
for i, variable in enumerate(['sepal_length', 'sepal_width'], start=1):
    plt.subplot(1, 2, i)

    datos = [
        df[df['species'] == especie][variable]
        for especie in df['species'].unique()
    ]

    plt.boxplot(datos, labels=df['species'].unique())
    plt.title(f'Distribucin de {variable} por especie')
    plt.xlabel('Especie')
    plt.ylabel(variable)

# d) Ajustar el layout para que no se encimen los grficos
plt.tight_layout()

# d) Mostrar el grfico
plt.show()

# e) Grfico de lneas
# e) Objetivo: Mostrar la tendencia de la longitud del ptalo (petal_length)
# e) a travs de las observaciones del dataset

plt.figure(figsize=(8, 5))

# e) Eje X: ndice de las observaciones
# e) Eje Y: longitud del ptalo
plt.plot(df.index, df['petal_length'])

# e) Etiquetas y ttulo
plt.xlabel('ndice de las observaciones')
plt.ylabel('Largo del ptalo (petal_length)')
plt.title('Tendencia del largo del ptalo a lo largo del dataset')

# e) Mostrar el grfico
plt.show()






```
