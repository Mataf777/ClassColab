#**Objetivo**
Utilizar plt.subplots() para visualizar diferentes distribuciones generadas con NumPy y almacenadas en un DataFrame.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt



np.random.seed(42)


#Generar 1000 valores para cada distribucin
normal = np.random.normal(loc=0, scale=1, size=1000)
uniforme = np.random.uniform(low=0, high=10, size=1000)
exponencial = np.random.exponential(scale=1, size=1000)
poisson = np.random.poisson(lam=3, size=1000)


#DataFrame
df = pd.DataFrame({
    'Normal': normal,
    'Uniforme': uniforme,
    'Exponencial': exponencial,
    'Poisson': poisson
})

df.head(1)

#Crear figura con cuadrcula 2x2
fig, ax = plt.subplots(2, 2, figsize=(10, 8))

#Histograma distribucin Normal
ax[0, 0].hist(df['Normal'], bins=30, edgecolor='black', linewidth=.5)
ax[0, 0].set_title('Distribucin Normal')

#Histograma distribucin Uniforme
ax[0, 1].hist(df['Uniforme'], bins=30, edgecolor='black', linewidth=.5)
ax[0, 1].set_title('Distribucin Uniforme')

#Histograma distribucin Exponencial
ax[1, 0].hist(df['Exponencial'], bins=30, edgecolor='black', linewidth=.5)
ax[1, 0].set_title('Distribucin Exponencial')

#Histograma distribucin Poisson
ax[1, 1].hist(df['Poisson'], bins=30, edgecolor='black', linewidth=.5)
ax[1, 1].set_title('Distribucin Poisson')


#Ajustar espacios automticamente

plt.suptitle("Graficas de distribuciones", fontsize=22, fontweight="bold")

plt.tight_layout()

plt.show()


```
