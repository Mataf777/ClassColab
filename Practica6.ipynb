```python
import pandas as pd


ruta = "/content/drive/MyDrive/ClassFiles/TitanicDataset.csv"
df = pd.read_csv(ruta)

# Agregado para conocer las dimensiones del set
print("Dimensiones del dataset:", df.shape)



# Calcular proporción de supervivencia por Sex y Pclass
survival_rates = (
    df
    .groupby(["Sex", "Pclass"], as_index=False)["Survived"]
    .mean()
    .rename(columns={"Survived": "survival_rate"})
)

print("Proporción de supervivencia por combinación (Sex, Pclass):")
display(survival_rates)

# Identificar la combinación con mayor tasa de supervivencia
max_rate = survival_rates["survival_rate"].max()
max_combination = survival_rates[survival_rates["survival_rate"] == max_rate]

print("== COMBINACIÓN CON MAYOR TASA DE SUPERVIVENCIA ==")
display(max_combination)

# Identificar la combinación con menor tasa de supervivencia
min_rate = survival_rates["survival_rate"].min()
min_combination = survival_rates[survival_rates["survival_rate"] == min_rate]

print("== COMBINACIÓN CON MENOR TASA DE SUPERVIVENCIA ==")
display(min_combination)

# Crear la columna FamilySize
df["FamilySize"] = df["SibSp"] + df["Parch"]

# Identificar pasajeros de familias grandes (FamilySize > 3)
familias_grandes = df[df["FamilySize"] > 3]

# Número de pasajeros en familias grandes
num_familias_grandes = len(familias_grandes)

print("Número de pasajeros en familias grandes:", num_familias_grandes)

# Proporción de supervivencia en familias grandes
proporcion_supervivencia_familias_grandes = familias_grandes["Survived"].mean()

print(
    f"Proporción de supervivencia en familias grandes:",
    f"{proporcion_supervivencia_familias_grandes:.2f}")

# Clasificar pasajeros por edad
def clasifica_edad(age):
    return "menor" if age < 18 else "mayor"

df["AgeGroup"] = df["Age"].apply(clasifica_edad)

conteo_edades = df["AgeGroup"].value_counts()

num_menores = conteo_edades.get("menor", 0)
num_mayores = conteo_edades.get("mayor", 0)

print("Menores de edad (<18):", num_menores)
print("Mayores de edad (>=18):", num_mayores)

# Promedios con NumPy vs Pandas
import numpy as np

# Promedios con NumPy
age_mean_np  = np.nanmean(df["Age"].to_numpy())
fare_mean_np = np.nanmean(df["Fare"].to_numpy())

# Promedios con Pandas
age_mean_pd  = df["Age"].mean(skipna=True)
fare_mean_pd = df["Fare"].mean(skipna=True)

# Mostrar resultados formateados a 2 decimales
print(f"Promedio Age - NumPy:  {age_mean_np:.2f} | Pandas: {age_mean_pd:.2f}")
print(f"Promedio Fare - NumPy: {fare_mean_np:.2f} | Pandas: {fare_mean_pd:.2f}")

# Verificar consistencia numérica (tolerancia por flotantes)
age_consistente  = np.allclose(age_mean_np, age_mean_pd, rtol=1e-12, atol=1e-12)
fare_consistente = np.allclose(fare_mean_np, fare_mean_pd, rtol=1e-12, atol=1e-12)

print("¿Age consistente entre NumPy y Pandas?:", age_consistente)
print("¿Fare consistente entre NumPy y Pandas?:", fare_consistente)


# Crear 5 intervalos equidistantes con NumPy
bins = np.linspace(df["Fare"].min(), df["Fare"].max(), 6)

# Asignar intervalo a cada pasajero
df["FareInterval"] = pd.cut(df["Fare"], bins=bins, include_lowest=True)

# Número de pasajeros por intervalo
print("Número de pasajeros por intervalo:")
print(df["FareInterval"].value_counts(sort=False))

# Proporción de supervivencia por intervalo
print("\nProporción de supervivencia por intervalo:")
print(df.groupby("FareInterval")["Survived"].mean())
```
