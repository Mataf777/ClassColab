#**Objetivo**
Utilizar GridSpec para organizar subplots de diferentes tamaños en un análisis de ventas, creando un dataframe con numeros aleatorios.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec


np.random.seed(42)


meses = ["Ene", "Feb", "Mar", "Abr", "May", "Jun",
         "Jul", "Ago", "Sep", "Oct", "Nov", "Dic"]

#ventas y gastos aleatorios
ventas = np.random.randint(2000, 5001, 12)
gastos = np.random.randint(1000, 4001, 12)

#DataFrame
df = pd.DataFrame({
    "Mes": meses,
    "Ventas": ventas,
    "Gastos": gastos
})

df


fig = plt.figure(figsize=(12, 8))

#GridSpec 2 filas x 2 columnas
gs = GridSpec(2,2, figure=fig)


# Primer gráfico (Barras) - ocupa toda la fila superior
ax1 = fig.add_subplot(gs[0, :])
ax1.bar(df["Mes"], df["Ventas"])
ax1.set_title("Ventas Mensuales")
ax1.set_ylabel("Ventas")

# Segundo gráfico (Línea) - parte inferior izquierda
ax2 = fig.add_subplot(gs[1, 0])
ax2.plot(df["Mes"], df["Gastos"], marker='o')
ax2.set_title("Gastos Mensuales")
ax2.set_ylabel("Gastos")

# Tercer gráfico (Dispersión) - ocupa dos filas en la columna derecha
ax3 = fig.add_subplot(gs[1,1])
ax3.scatter(df["Ventas"], df["Gastos"])
ax3.set_title("Relación Ventas vs Gastos")
ax3.set_xlabel("Ventas")
ax3.set_ylabel("Gastos")

# Ajustar espacios

plt.suptitle("Analisis de Ventas", fontsize=22, fontweight="bold")
plt.tight_layout()
plt.show()

```
