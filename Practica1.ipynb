# Practica 1
1) En esta práctica se analiza el archivo mbox.txt para identificar y extraer únicamente los valores numéricos asociados a la cadena X-DSPAM-Confidence. El código recorre cada línea del archivo, selecciona aquellas que comienzan con dicha cadena mediante la instrucción startswith, separa el texto para obtener el número y lo guarda en una lista. Posteriormente, se muestran todas las mediciones encontradas y se calculan estadísticas básicas: la cantidad de datos extraídos, la sumatoria de los valores, el promedio y la varianza. De esta manera, se refuerza el uso de estructuras de control, listas y operaciones matemáticas en Python dentro del entorno de Google Colab, aplicando conceptos vistos en el laboratorio previo.




```python


# Parámetros
RUTA_ARCHIVO = '/content/drive/MyDrive/ClassFiles/mbox.txt'


valores = []

try:

    with open(RUTA_ARCHIVO, 'r', encoding='utf-8', errors='ignore') as f:
        for linea in f:
            # 1) Filtrar solo líneas que comienzan con el prefijo requerido
            if linea.startswith('X-DSPAM-Confidence:'):

                try:
                    numero = float(linea.split(':', 1)[1].strip())
                except (ValueError, IndexError):

                    continue

                valores.append(numero)

                # 2) Mostrar el resultado de cada medición encontrada
                print(f"Medición encontrada: {numero}")

    # 3) Cálculos: cantidad, suma, promedio, varianza (poblacional)
    n = len(valores)
    suma = sum(valores) if n > 0 else 0.0
    promedio = (suma / n) if n > 0 else 0.0

    # Varianza poblacional: divide entre n
    if n > 0:
        varianza_poblacional = sum((x - promedio) ** 2 for x in valores) / n
    else:
        varianza_poblacional = 0.0

    print("\n--- Resumen ---")
    print(f"Cantidad de datos extraídos: {n}")
    print(f"Suma de los datos: {suma:.6f}")
    print(f"Promedio de los datos: {promedio:.6f}")
    print(f"Varianza (poblacional): {varianza_poblacional:.6f}")

except FileNotFoundError:
    print(f"Error: no se encontró el archivo en la ruta: {RUTA_ARCHIVO}")
except PermissionError:
    print("Error: permisos insuficientes para leer el archivo.")
except UnicodeDecodeError:
    print("Error de decodificación: prueba con otro encoding (p. ej., 'latin-1').")
except Exception as e:
    print(f"Ocurrió un error inesperado: {type(e).__name__}: {e}")
```
