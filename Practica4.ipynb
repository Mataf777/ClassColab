```python
import numpy as np


# Matrices 3x3 con valores aleatorios entre 1 y 10
matriz1 = np.random.randint(1, 11, size=(3, 3))
matriz2 = np.random.randint(1, 11, size=(3, 3))

print("Matriz 1:")
print(matriz1)

print("\nMatriz 2:")
print(matriz2)

# Suma de matrices
suma = matriz1 + matriz2

# Resta de matrices
resta = matriz1 - matriz2

# Multiplicaci贸n de la primera matriz por 2
multiplicacion_escalar = matriz1 * 2

# Multiplicaci贸n matricial
multiplicacion_matricial = matriz1 @ matriz2

print("\nSuma de matrices:")
print(suma)

print("\nResta (matriz1 - matriz2):")
print(resta)

print("\nMatriz 1 multiplicada por 2:")
print(multiplicacion_escalar)

print("\nMultiplicaci贸n matricial (matriz1 @ matriz2):")
print(multiplicacion_matricial)
```
