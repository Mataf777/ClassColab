#**Objetivo**

Utilizar Seaborn junto con plt.subplots() para visualizar diferentes aspectos de un conjunto de datos en una sola figura, usando dataset "penguins" de Seaborn

```python

import seaborn as sns
import matplotlib.pyplot as plt


df = sns.load_dataset("penguins")


sns.set(style="whitegrid", context="notebook")

#Crear figura y ejes 2x2
fig, axes = plt.subplots(2, 2, figsize=(12, 10))


palette_species = "Set2"
hist_color = "#4C72B0"  # azul suave

#(0,0): Histograma de longitudes de aletas
sns.histplot(
    data=df,
    x="flipper_length_mm",
    bins=20,
    kde=False,
    color=hist_color,
    ax=axes[0, 0]
)
axes[0, 0].set_title("Histograma: Longitud de aleta (mm)")
axes[0, 0].set_xlabel("Longitud de aleta (mm)")
axes[0, 0].set_ylabel("Frecuencia")

#(0,1): Boxplot comparando masa corporal por especie

sns.boxplot(
    data=df,
    x="species",
    y="body_mass_g",
    hue="species",
    dodge=False,
    palette=palette_species,
    ax=axes[0, 1],
    legend=False
)
axes[0, 1].set_title("Boxplot: Masa corporal por especie")
axes[0, 1].set_xlabel("Especie")
axes[0, 1].set_ylabel("Masa corporal (g)")


#(1,0): Dispersin entre masa corporal y largo de aletas
sns.scatterplot(
    data=df,
    x="flipper_length_mm",
    y="body_mass_g",
    hue="species",
    palette=palette_species,
    s=60,
    alpha=0.8,
    ax=axes[1, 0]
)
axes[1, 0].set_title("Dispersin: Masa corporal vs Longitud de aleta")
axes[1, 0].set_xlabel("Longitud de aleta (mm)")
axes[1, 0].set_ylabel("Masa corporal (g)")
axes[1, 0].legend(title="Especie", loc="best")

#(1,1): Violin plot de la masa corporal segn la especie

sns.violinplot(
    data=df,
    x="species",
    y="body_mass_g",
    hue="species",
    dodge=False,
    palette=palette_species,
    inner="quartile",
    ax=axes[1, 1],
    legend=False
)
axes[1, 1].set_title("Violin plot: Masa corporal por especie")
axes[1, 1].set_xlabel("Especie")
axes[1, 1].set_ylabel("Masa corporal (g)")


# 7) Ajustar diseo
plt.suptitle("Graficas de Pinginos", fontsize=22, fontweight="bold")
plt.tight_layout()
plt.show()
```
