```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt


# dataset "tips" desde seaborn
df = sns.load_dataset('tips')


# Exploracin inicial del dataset
print(df.head())


# Informacin general del dataset
# Columnas, tipos de datos, valores no nulos
df.info()


# Estadsticas descriptivas de las variables numricas
df.describe()


# Verificar la existencia de valores nulos
df.isnull().sum()



# a) Mapa de Calor (Heatmap)
# a) Objetivo: Visualizar la correlacin entre variables numricas

# a) Seleccionar variables numricas de inters
variables_numericas = df[['total_bill', 'tip', 'size']]

# a) Calcular la matriz de correlacion
correlacion = variables_numericas.corr()

# a) Crear el mapa de calor
plt.figure(figsize=(6, 5))
sns.heatmap(
    correlacion,
    annot=True,
    cmap='coolwarm',
    fmt='.2f'
)

plt.title('Mapa de calor de correlacion (Tips)')
plt.show()



# b) Diagrama de Violin
# b) Objetivo: Distribucin de las propinas por dia de la semana


plt.figure(figsize=(8, 5))
sns.violinplot(
    x='day',
    y='tip',
    data=df,
    palette='pastel'
)

plt.xlabel('Da de la semana')
plt.ylabel('Propina (tip)')
plt.title('Distribucin de las propinas por dia')
plt.show()


# c) Grfico de dispersin (Scatter Plot)
# c) Objetivo: Relacin entre total_bill y tip


plt.figure(figsize=(8, 5))
sns.scatterplot(
    x='total_bill',
    y='tip',
    hue='day',
    data=df
)

plt.xlabel('Total de la cuenta (total_bill)')
plt.ylabel('Propina (tip)')
plt.title('Relacin entre total de la cuenta y propina')
plt.legend(title='Da')
plt.show()



# d) Grfico de barras (Bar Plot)
# d) Objetivo: Promedio de la propina por gnero


plt.figure(figsize=(6, 5))
sns.barplot(
    x='sex',
    y='tip',
    data=df,
    estimator='mean',
    palette='muted'
)

plt.xlabel('Gnero')
plt.ylabel('Propina promedio')
plt.title('Promedio de la propina por gnero')
plt.show()
```
